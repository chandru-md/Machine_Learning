```python
df_france=pd.pivot_table(data=france_df,values="Quantity",index="InvoiceNo",columns="Description").fillna(0)
```

```python
df_france
```

```python
df_france=df_france.applymap(lambda x:1 if x>0 else 0)
```

```python
pip install mlxtend
```

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
frequent_item=apriori(df=df_france,min_support=0.05,use_colnames=True)
frequent_item
```

```python
best_association=association_rules(frequent_item)
best_association
```
