```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x=credit_card_df.drop(labels="DEFAULT",axis=1)
```

```python
x
```

```python
y=credit_card_df["DEFAULT"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
RF_model=RandomForestClassifier()
AG_model=AdaBoostClassifier()
G_model=GradientBoostingClassifier()
XGB_model=XGBClassifier()
LGB_model=LGBMClassifier()
```

```python
%%time
RF_model.fit(x_train,y_train)
```

```python
y_pred=RF_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
AG_model.fit(x_train,y_train)
```

```python
y_pred=AG_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
G_model.fit(x_train,y_train)
```

```python
y_pred=G_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
XGB_model.fit(x_train,y_train)
```

```python
y_pred=XGB_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
%%time
LGB_model.fit(x_train,y_train)
```

```python
y_pred=LGB_model.predict(x_test)
accuracy_score(y_test,y_pred)
```
