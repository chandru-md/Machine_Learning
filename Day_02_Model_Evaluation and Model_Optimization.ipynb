# AIC = the akaike information criterion (AIC)
## the akaike information criterion (AIC) is a metric used to compare the fit of different regression models
## The lower AIC is better the model fit.

# BIC = Bayesian  information criterion (BIC)
## Bayesian  information criterion (BIC) is a metric used to compare the goodness of fit of different regression models
## the lower BIC indicates a better balance between model fit and complexity

# R sqaured = 1-(RSS/TSS)
## Coefficient of determination
## RSS- sum of squares od residuals
## TSS- total sum of squares
## Higher the Rsquared the better the model.

# Adjusted R Sqaured 
## is a modified version of R-Squared that adjusts for predictors that are not significant in a regression model
## a lower Adjusted r ^2 indicates that the additional input variables are not adding value to the model
## The higher the rsqaured_adj the better- model fit.
