```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#select from model
#Recursive feature elimination
```

```python
data=load_breast_cancer()
```

```python
data
```

```python
df=pd.DataFrame(data=data.data,columns=data.feature_names)
```

```python
df
```

```python
y=data.target
```

```python
x=df
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
x.dtypes
```

```python
x.isna().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
Select_From_Model=SelectFromModel(estimator=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=None))
```

```python
Select_From_Model.fit(x_train,y_train)
```

```python
Select_From_Model.get_support() #its giving me which features is important.
```

```python
x.columns[Select_From_Model.get_support()]
```

```python
x_train_transformed=Select_From_Model.transform(x_train)
```

```python
x_train_transformed
```

```python
x_train_transformed.shape
```

```python
x_test_transformed=Select_From_Model.transform(x_test)
```

```python
x_test_transformed.shape
```

```python
def run_Randomforest(x_train,x_test,y_train,y_test):
    rf_model=RandomForestClassifier(max_depth=7)
    rf_model.fit(x_train,y_train)
    y_pred=rf_model.predict(x_test)
    print("accuracy_score:",accuracy_score(y_test,y_pred))
    print("confusion_matrix:",confusion_matrix(y_test,y_pred))
```

```python
x_train
```

```python
x_train_transformed.shape
```

```python
x_test_transformed.shape
```

```python
%%time
run_Randomforest(x_train,x_test,y_train,y_test)
```

```python
%%time
run_Randomforest(x_train_transformed,x_test_transformed,y_train,y_test)
```

```python
#RFE = Recursive feature elimination
```

```python
from sklearn.feature_selection import RFE
```

```python
RFE_model=RFE(estimator=RandomForestClassifier(n_estimators=50,criterion='gini',max_depth=10))
RFE_model.fit(x_train,y_train)
```

```python
RFE_model.get_support()
```

```python
x.columns[RFE_model.get_support()]
```

```python
x_train_transformed_rfe=RFE_model.transform(x_train)
x_test_transformed_rfe=RFE_model.transform(x_test)
```

```python
%%time
run_Randomforest(x_train_transformed_rfe,x_test_transformed_rfe,y_train,y_test)
```

```python
for i in range(1,len(x.columns)):
    RFE_model=RFE(estimator=RandomForestClassifier(n_estimators=50,criterion='gini',max_depth=10),n_features_to_select=i)
    RFE_model.fit(x_train,y_train)
    x_train_transformed_rfe=RFE_model.transform(x_train)
    x_test_transformed_rfe=RFE_model.transform(x_test)
    print(x_train_transformed_rfe.shape)
    print("Num of feature",i)
    run_Randomforest(x_train_transformed_rfe,x_test_transformed_rfe,y_train,y_test)

    print("###################################")
```
